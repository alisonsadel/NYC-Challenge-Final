```python
import pandas as pd

# Remove dataFrame display size restrictions
#pd.set_option("display.max_rows", None, "display.max_columns", None)

housing = pd.read_csv("output/re_data_clean.csv")
walk = pd.read_csv("output/walkscore_data_clean.csv")
trees = pd.read_csv("output/tree_data_clean.csv")
stations = pd.read_csv("output/stations_data_clean.csv")
brownfields = pd.read_csv("output/brownfields_data_clean.csv")
crime = pd.read_csv("output/crime_data_clean.csv")
elem = pd.read_csv("output/final_elementary.csv")
middle = pd.read_csv("output/final_middle.csv")
high = pd.read_csv("output/final_high.csv")
```

### Walkability

```python
walk = walk[['zipcode', 'walkable_false', 'walkable_true']]

walk = walk.groupby(by=["zipcode"]).sum()

walk.head()
```

```python
trees.columns
```

### Trees

```python
trees = trees[['zipcode','health_level_dead',
       'health_level_fair', 'health_level_good', 'health_level_poor']] 

trees = trees.groupby(by=["zipcode"]).sum()

trees.head()
```

```python
stations.columns
```

### Stations

```python
stations = stations[['zipcode','ada_access_no',
       'ada_access_yes', 'ada_access_partial']]


stations = stations.groupby(by=["zipcode"]).sum()

stations.head()
```

### Brownfields

```python
brownfields = brownfields[['zipcode','type_is_active_cleanup_mild_threat',
       'type_is_active_significant_threat', 'type_is_completed_clean',
       'type_is_maintain_continued_threat', 'type_is_potential_threat']]


brownfields = brownfields.groupby(by=["zipcode"]).sum()

brownfields.head()
```

### Crime

```python

crime = crime[['zipcode', 'type_is_felony', 'type_is_violation', 'type_is_misdemeanor', 'danger_level_is_high', 'danger_level_is_low']]


crime = crime.groupby(by=["zipcode"]).sum()


crime.head()

```
